```python
# Classify movies based on ratings [excellent, good, average]
```

```python
data.columns
```

```python
def rating(rating):
    if rating>=7.0:
        return "Excellent"
    elif rating>=6.0:
        return "Good"
    else:
        return "Average"
```

```python
data['Rating_movie']=data['Rating'].apply(rating)
```
